Develop a guessing game that performs the following:

1. The system will automatically generate a random number between 1 and 100.

2. Users can insert any number between 1 and 100

3. The program shall be able to compare the number generated by the system and the number that has been entered by the user. The program shall print out one of the following options to help the user improve the next guess:
```
   • You are right, great job!
   • Your guess is low, try again!
   • your guess is high, try again!
```

The program exits when the user guess matches the number generated by the system

````python
import random


random_numbers = random.randint(1,101)
print (random_numbers)
````

````python
user_input = int(input("Enter your guess between 1 and 100: "))
print("Invalid input. Please enter a valid integer.")

while True:
  if random_numbers == user_input:
    print ("You're Right! Good job")
    break
  elif random_numbers > user_input:
    print ("Your guess is low, try again!")
    user_input = int(input("Enter your guess between 1 and 100: "))
  elif random_numbers < user_input:
    print ("Your guess is high, try again!")
    user_input = int(input("Enter your guess between 1 and 100: "))
````
